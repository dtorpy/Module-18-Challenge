```python
#import data before and after covid hit NYC in March 2020
import pandas as pd

Feb_df = pd.read_csv('Data/202002-citibike-tripdata.csv')
March_df = pd.read_csv('Data/202003-citibike-tripdata.csv')
April_df = pd.read_csv('Data/202004-citibike-tripdata.csv')
```

```python
#drop any null data
Feb_df = Feb_df.dropna()
March_df = March_df.dropna()
April_df = April_df.dropna()
```

```python
df2=Feb_df.sample(frac=0.01, replace=True, random_state=1)
df3=March_df.sample(frac=0.01, replace=True, random_state=1)
df4=April_df.sample(frac=0.01, replace=True, random_state=1)
```

```python
df2.count()
```

```python
df3.count()
```

```python
df4.count()
```

```python
frames = [df2, df3, df4]

result = pd.concat(frames)
```
